```python
from collections import defaultdict
category = defaultdict(list)

for index, row in categorys.iterrows():

    cat = row['categorys'].split("|")
    for c in cat:
        category['id'].append(row['id'])
        category['category'].append(c)
category = pd.DataFrame(category)
category
```

```python
year = anime[['id', 'year']]
year
```

```python
import re
yea = defaultdict(list)

for index, row in year.iterrows():
    row['year']= row['year'].replace('+',',')
    years = row['year'].split(",")
    for y in years:
        yea['id'].append(row['id'])
        yea['year'].append(y)
year = pd.DataFrame(yea)
year
```

```python
year['year'].unique()
```

```python
import re
yea = defaultdict(list)
for index, row in year.iterrows():
    row['year'] = row['year'].replace('tháng ', '')
    row['year'] = re.split('/', row['year'])[-1]
    if row['year']=='Đang cập nhật':
        row['year'] = 0
    if row['year']=='042020':
        row['year']= row['year'][2:]
    yea['id'].append(row['id'])
    yea['year'].append(int(row['year']))

year = pd.DataFrame(yea)
year
```
